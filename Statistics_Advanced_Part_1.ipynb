13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem states that:

* When you take sufficiently large random samples from any population (regardless of its distribution), the sampling distribution of the sample mean will approach a normal distribution.

* Key Points:
Applies to sample means (or sums) from independent, identically distributed random variables.

* The larger the sample size
n, the closer the distribution of the sample mean to a normal distribution.

14. How does the Central Limit Theorem relate to the normal distribution?

* The Central Limit Theorem (CLT) directly links to the normal distribution by stating that:

1. Regardless of the shape of the original population distribution,

2. The distribution of the sample mean (or sum) from large enough random samples will approximate a normal distribution.

In other words, the CLT explains why the normal distribution is so prevalent in statistics: it emerges naturally as the distribution of averages when sampling from almost any population.

This relationship allows statisticians to apply normal distribution methods (like confidence intervals and hypothesis tests) to sample means even when the original data isn’t normal.

15. What is the application of Z statistics in hypothesis testing?

Steps in Hypothesis Testing Using Z-Statistic

1. **Formulate Hypotheses:**
   - Null hypothesis: $H_0: \mu = \mu_0$
   - Alternative hypothesis: $H_a: \mu \ne \mu_0$ (or $<,\ >$)

2. **Calculate the Z-Statistic:**

   $$
   Z = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}}
   $$

   Where:
   - $\bar{X}$ = Sample mean  
   - $\mu_0$ = Hypothesized population mean  
   - $\sigma$ = Population standard deviation  
   - $n$ = Sample size

3. **Find the Critical Z-Value:**
   - Based on the significance level $\alpha$ (e.g., 0.05), find the critical value(s) from the standard normal distribution.

4. **Make a Decision:**
   - If $|Z| > Z_{\text{critical}}$, **reject $H_0$**
   - Otherwise, **fail to reject $H_0$**



16. How do you calculate a Z-score, and what does it represent?

A Z-score indicates how many standard deviations a data point is from the mean.

Formula:

$$
Z = \frac{x - \mu}{\sigma}
$$

Where:  
- $x$ = data point  
- $\mu$ = population mean  
- $\sigma$ = population standard deviation


### 📌 Interpretation:

- $Z = 0$: Value is at the mean  
- $Z > 0$: Value is above the mean  
- $Z < 0$: Value is below the mean  
- $Z = 2$: Value is 2 standard deviations above the mean



17. What are point estimates and interval estimates in statistics?

- Example: The sample mean $\bar{x}$ is a point estimate of the population mean $\mu$.


**Interval Estimate:**

An **interval estimate** provides a range (confidence interval) within which the parameter is likely to fall.

- Example: "We are 95% confident that the population mean lies between 72 and 78."


| Feature              | Point Estimate             | Interval Estimate                          |
|----------------------|-----------------------------|---------------------------------------------|
| Definition           | Single value estimate       | Range of values (confidence interval)       |
| Example              | $\bar{x} = 75$              | 95% CI: (72, 78)                            |
| Reliability          | Less reliable               | More reliable                               |

22. What is the concept of expected value in a probability distribution?

The **expected value (E[X])** is the theoretical mean of a random variable.

* For a Discrete Variable:
$$
E(X) = \sum_{i=1}^{n} x_i \cdot P(x_i)
$$

* For a Continuous Variable:
$$
E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
$$

The expected value gives the **average outcome** if the experiment is repeated many times.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_z_scores(data):
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 4))
    plt.scatter(range(len(z_scores)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title('Z-scores Plot')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage
data = [10, 12, 9, 15, 18, 20, 21, 19, 14, 13]
z_scores = plot_z_scores(data)
print("Z-scores:", z_scores)
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters
population_size = 100000
sample_sizes = [1, 5, 30, 100]  # Different sample sizes to test CLT
num_samples = 1000  # Number of samples to draw for each sample size

# Generate population from a non-normal distribution (exponential)
population = np.random.exponential(scale=1.0, size=population_size)

plt.figure(figsize=(12, 10))

for i, n in enumerate(sample_sizes, 1):
    sample_means = []

    # Draw num_samples samples of size n and compute their means
    for _ in range(num_samples):
        sample = np.random.choice(population, size=n, replace=False)
        sample_means.append(np.mean(sample))
    sample_means = np.array(sample_means)

    # Plot histogram of sample means
    plt.subplot(2, 2, i)
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')

    # Plot normal distribution fit for sample means
    mu, sigma = np.mean(sample_means), np.std(sample_means)
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)
    plt.plot(x, stats.norm.pdf(x, mu, sigma), 'r-', lw=2, label='Normal PDF')

    plt.title(f'Sample size = {n}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulate_binomial_distribution(n=10, p=0.5, size=1000):
    # Generate random variables from binomial distribution
    data = np.random.binomial(n=n, p=p, size=size)

    # Get unique values and their frequencies
    values, counts = np.unique(data, return_counts=True)

    # Convert frequencies to probabilities
    empirical_probs = counts / size

    # Calculate theoretical probabilities using binomial PMF
    x = np.arange(0, n + 1)
    theoretical_probs = binom.pmf(x, n, p)

    # Plot empirical vs. theoretical probabilities
    plt.figure(figsize=(8, 5))
    plt.bar(values, empirical_probs, alpha=0.6, label='Empirical Probabilities', color='skyblue', edgecolor='black')
    plt.plot(x, theoretical_probs, 'ro-', label='Theoretical PMF', linewidth=2)

    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
simulate_binomial_distribution(n=10, p=0.5, size=1000)
```

19. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test.
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Decision
    reject_null = p_value < alpha

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.2f}")
    print(f"P-value: {p_value:.4f}")
    print("Result:", "Reject H₀" if reject_null else "Fail to Reject H₀")

    return z_stat, p_value, reject_null

# Example usage
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=50)  # Sample from N(102, 15)
population_mean = 100
population_std = 15  # Assume known population std deviation

z_test(sample, population_mean, population_std, alpha=0.05)
```

20. Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.

    Parameters:
        data (list or array): The dataset
        confidence (float): Confidence level (default is 0.95 for 95%)

    Returns:
        (lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")

    return lower, upper

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=10, size=30)  # Sample from N(100, 10)
confidence_interval(sample_data, confidence=0.95)
```

```python
import numpy as np
from scipy import stats

# Step 1: Generate data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Std=10, n=100

# Step 2: Calculate 95% confidence interval for the mean
mean = np.mean(data)
std_err = stats.sem(data)  # Standard error of the mean
confidence = 0.95
n = len(data)

# t critical value
t_critical = stats.t.ppf((1 + confidence) / 2, df=n - 1)

# Margin of error
margin_of_error = t_critical * std_err

# Confidence Interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Step 3: Display results
print(f"Sample Mean: {mean:.2f}")
print(f"Standard Error: {std_err:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

22. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate x values (range of values for which PDF will be calculated)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the PDF values
pdf = norm.pdf(x, mean, std_dev)

# Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label=f'N({mean}, {std_dev}²)', color='blue')
plt.title('Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define Poisson distribution parameter (λ)
lambda_param = 4

# Generate x values (discrete range)
x = np.arange(0, 15)

# Calculate CDF values
cdf_values = poisson.cdf(x, mu=lambda_param)

# Print a few values for interpretation
for k, cdf in zip(x, cdf_values):
    print(f"P(X ≤ {k}) = {cdf:.4f}")

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_values, where='post', label=f'Poisson(λ={lambda_param})', color='green')
plt.title('CDF of Poisson Distribution')
plt.xlabel('k (Number of events)')
plt.ylabel('P(X ≤ k)')
plt.grid(True)
plt.legend()
plt.show()
```

25. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two sample datasets
np.random.seed(0)
data1 = np.random.normal(loc=50, scale=5, size=1000)   # Dataset 1: mean=50, std=5
data2 = np.random.normal(loc=50, scale=10, size=1000)  # Dataset 2: mean=50, std=10

# Calculate standard deviations
std1 = np.std(data1, ddof=1)  # ddof=1 for sample std deviation
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Visualize the datasets with histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data1, bins=30, color='blue', alpha=0.7)
plt.title(f'Dataset 1 (Std Dev: {std1:.2f})')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(data2, bins=30, color='orange', alpha=0.7)
plt.title(f'Dataset 2 (Std Dev: {std2:.2f})')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.suptitle('Comparison of Standard Deviations')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    skewness = skew(data)
    kurt = kurtosis(data)  # By default, Fisher’s definition (subtract 3)
    return skewness, kurt

# Generate sample data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate
skewness_val, kurtosis_val = calculate_skewness_kurtosis(data)

print(f"Skewness: {skewness_val:.4f}")
print(f"Kurtosis: {kurtosis_val:.4f}")
```
