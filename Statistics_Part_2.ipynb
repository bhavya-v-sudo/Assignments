# Statistics Part 2

1. What is hypothesis testing in statistics?

* Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It allows us to test an assumption (called a hypothesis) about a population parameter.



2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

* The null hypothesis (H₀) and the alternative hypothesis (H₁ or Ha) are two opposing statements in hypothesis testing.

*  Null Hypothesis (H₀):
  * It is the default assumption or starting point.
  * It is assumed to be true unless there is strong evidence against it.

* Alternative Hypothesis (H₁ or Ha):
  * It is what you are trying to prove or support.
  * It is considered only if the null hypothesis is rejected.



3. What is the significance level in hypothesis testing, and why is it important?

* The significance level, denoted by α (alpha), is the threshold probability used to decide whether to reject the null hypothesis (H₀) in a hypothesis test.

* It represents the maximum probability you're willing to accept for making a Type I error, which is rejecting H₀ when it's actually true.

* It's the cut-off to compare with the p-value:

  * If p-value ≤ α → reject H₀ (significant result)

  * If p-value > α → fail to reject H₀ (not significant)

4. What does a P-value represent in hypothesis testing?

* The P-value is the probability of obtaining a test statistic (like a sample mean or proportion) as extreme as or more extreme than the one observed, assuming that the null hypothesis is true.

* A small P-value (typically ≤ 0.05) means:
  * The observed data is unlikely under the assumption that H₀ is true.
  * Reject Ho

* A large P-value (typically > 0.05) means:
  * The observed data is likely under H₀.
  * Fail to reject Ho

5. How do you interpret the P-value in hypothesis testing?

* A small P-value indicates the observed data is unlikely under H₀ → strong evidence against H₀.

* A large P-value means the data is likely under H₀ → do not reject H₀.

6. What are Type 1 and Type 2 errors in hypothesis testing?

* Type I Error (False Positive)
  * You reject the null hypothesis (H₀) even though it is actually true.
  * Probability of Type I error = α (significance level).

* Type II Error (False Negative)
  * You fail to reject the null hypothesis when the alternative hypothesis is true.
  * Probability of Type II error = β

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

* One-Tailed Test:
  * Tests for an effect in one specific direction.
  * You are only interested in whether a parameter is greater than or less than a certain value, not both.
  * Examples:
    * H₀: μ = 100
    * H1: μ > 100 -> Right-tailed test
    * H1: μ < 100 -> Left-tailed test

* Two-Tailed Test:
  * Tests for an effect in either direction — higher or lower.
  * You're checking if a parameter is different from a certain value, in either direction.
  * Examples:
    * H₀: μ = 100
    * H₁: μ ≠ 100


8. What is the Z-test, and when is it used in hypothesis testing?

* A Z-test is a statistical test used to determine whether there is a significant difference between a sample mean and a population mean, or between two proportions, when certain conditions are met.

* The sample size is large (usually n ≥ 30),
OR the population is normally distributed.

* The population standard deviation (σ) is known.

*  You are testing the mean or proportion.


9. How do you calculate the Z-score, and what does it represent in hypothesis testing?


**Formula for Z-score (One-Sample Z-test for Mean)**

$$
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
$$

Where:  
- $\bar{x}$ = sample mean  
- $\mu$ = population mean under $H_0$  
- $\sigma$ = population standard deviation  
- $n$ = sample size

- The **Z-score** tells you how far, in standard error units, your sample result is from the expected value under the null hypothesis.
- It follows a **standard normal distribution** (mean = 0, standard deviation = 1).

**Example**

Suppose:  
- Sample mean $\bar{x} = 105$  
- Population mean $\mu = 100$  
- Population standard deviation $\sigma = 10$  
- Sample size $n = 25$

Then,

$$
Z = \frac{105 - 100}{10 / \sqrt{25}} = \frac{5}{2} = 2.5
$$

Since $Z = 2.5$ is greater than $1.96$, you **reject the null hypothesis** at $\alpha = 0.05$ (two-tailed test).


10. What is the T-distribution, and when should it be used instead of the normal distribution?

The T-distribution (also called Student's t-distribution) is a probability distribution used in hypothesis testing and confidence intervals when the sample size is small and/or the population standard deviation ($\sigma$) is unknown.

Use the T-distribution when:

1. The sample size is small (typically $n < 30$)
2. The population standard deviation ($\sigma$) is unknown
3. The underlying population is normally distributed (or approximately normal)

11. What is the difference between a Z-test and a T-test?

- Use a Z-test when:
  - You know the population standard deviation ($\sigma$)
  - You have a large sample size
  
- Use a T-test when:
  - You do not know $\sigma$
  - You have a small sample size
  - More robust when working with real-world sample data

12. What is the T-test, and how is it used in hypothesis testing?

* A **T-test** is a statistical test used to **compare means** and determine whether the difference between sample data and a population or between two samples is **statistically significant**, especially when the **population standard deviation is unknown** and/or the **sample size is small**.

* Steps in Hypothesis Testing with T-test

1. State hypotheses
   - $H_0$: no difference (e.g., $\mu = 100$)  
   - $H_1$: there is a difference (e.g., $\mu \neq 100$)

2. Choose significance level ($\alpha$)
   - Common values: 0.05, 0.01

3. Calculate t-statistic

4. Find critical t-value from the t-distribution table (based on $\alpha$ and df)

5. Compare
   - If $|t| > t_{\text{critical}}$: **Reject $H_0$**

6. Interpret
   - A significant result means evidence **against** the null hypothesis

13. What is the relationship between Z-test and T-test in hypothesis testing?


* Both the Z-test and T-test are statistical methods used in hypothesis testing to determine whether a sample mean significantly differs from a population mean or another sample mean.


* Similarities

  - Both are used to test hypotheses about means
  - Both compare an observed statistic (e.g., sample mean) to a hypothesized population mean
  - Both involve computing a test statistic and comparing it to a critical value
  - Both assume the underlying population is normally distributed (or approximately normal)


* Key Differences

| Feature                         | **Z-test**                                     | **T-test**                                      |
|----------------------------------|------------------------------------------------|--------------------------------------------------|
| Use case                        | Population standard deviation ($\sigma$) is **known** | Population standard deviation ($\sigma$) is **unknown** |
| Sample size                     | Typically used when **n ≥ 30**                 | Preferred when **n < 30**                        |
| Distribution used               | **Standard normal distribution (Z)**          | **T-distribution** with **degrees of freedom (df)** |
| Tail behavior                   | **Thinner tails**                             | **Heavier tails** (to account for more variability) |

* Conclusion

- Use Z-test when you have a large sample and know the population standard deviation
- Use T-test when the sample is small and σ is unknown
- They are fundamentally based on the same logic, but the T-test is more flexible in real-world scenarios



14. What is a confidence interval, and how is it used to interpret statistical results?


- **Confidence Level:** The percentage of times (e.g., 95%) that the confidence interval would contain the true population parameter if you repeated the study many times.

Formula for a Confidence Interval for a Mean:

When the population standard deviation ($\sigma$) is known (Z-distribution):

$$
\text{CI} = \bar{x} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
$$

When $\sigma$ is unknown and sample size is small (T-distribution):

$$
\text{CI} = \bar{x} \pm t_{\alpha/2, df} \times \frac{s}{\sqrt{n}}
$$

Where:  
- $\bar{x}$ = sample mean  
- $Z_{\alpha/2}$ = critical Z-value for confidence level (e.g., 1.96 for 95%)  
- $t_{\alpha/2, df}$ = critical t-value for confidence level and degrees of freedom  
- $\sigma$ = population standard deviation  
- $s$ = sample standard deviation  
- $n$ = sample size  
- $df$ = degrees of freedom ($n-1$)


**Example**

Sample mean test score: 80  
Sample standard deviation: 10  
Sample size: 25  
Confidence level: 95% ($t_{0.025,24} \approx 2.064$)

$$
\text{CI} = 80 \pm 2.064 \times \frac{10}{\sqrt{25}} = 80 \pm 2.064 \times 2 = 80 \pm 4.128
$$

**Confidence Interval: (75.87, 84.13)**

We are 95% confident that the true population mean lies between 75.87 and 84.13.



15. What is the Margin of Error, and How Does It Affect the Confidence Interval?


**Margin of Error (MoE)**

The margin of error is the maximum amount by which the estimated value (e.g., sample mean) is expected to differ from the true population parameter due to random sampling variability.

- It defines how precise your estimate is.
- It is half the width of the confidence interval.


**Formula for Margin of Error**

When the population standard deviation ($\sigma$) is known:

$$
\text{MoE} = Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
$$

When $\sigma$ is unknown and the sample size is small:

$$
\text{MoE} = t_{\alpha/2, df} \times \frac{s}{\sqrt{n}}
$$

Where:  
- $Z_{\alpha/2}$ = critical Z-value for the confidence level (e.g., 1.96 for 95%)  
- $t_{\alpha/2, df}$ = critical t-value for the confidence level and degrees of freedom  
- $\sigma$ = population standard deviation  
- $s$ = sample standard deviation  
- $n$ = sample size  
- $df$ = degrees of freedom ($n - 1$)




- The confidence interval is calculated as:

  $$
  \text{CI} = \text{Estimate} \pm \text{Margin of Error}
  $$

- A larger margin of error means a wider confidence interval, indicating less precision.
- A smaller margin of error means a narrower confidence interval, indicating more precision.




16. How is Bayes' Theorem Used in Statistics, and What is Its Significance?


Bayes' Theorem is a fundamental rule in probability theory that describes how to update the probability of a hypothesis based on new evidence or data.

It mathematically relates prior beliefs with likelihood of observed data to compute the posterior probability (updated belief).


Bayes' Theorem Formula

$$
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
$$

Where:  
- $P(A|B)$ = **Posterior probability**: Probability of hypothesis $A$ given evidence $B$  
- $P(B|A)$ = **Likelihood**: Probability of evidence $B$ given hypothesis $A$  
- $P(A)$ = **Prior probability**: Initial probability of hypothesis $A$ before seeing evidence  
- $P(B)$ = **Marginal likelihood**: Total probability of evidence $B$



Significance of Bayes' Theorem

- **Combines prior knowledge and evidence:** Makes use of existing information rather than treating all data as equally new.
- **Handles uncertainty naturally:** Provides full probability distributions for unknowns, not just point estimates.



17. What is the Chi-square Distribution, and When Is It Used?

If $Z_1, Z_2, ..., Z_k$ are independent standard normal variables, then:

$$
\chi^2 = \sum_{i=1}^k Z_i^2
$$

follows a Chi-square distribution with $k$ degrees of freedom.

- It is the distribution of a sum of the squares of independent standard normal random variables.
- It is asymmetric and skewed right, with shape depending on degrees of freedom (df).
- As the degrees of freedom increase, the distribution approaches a normal distribution.


* The Chi-square distribution is essential in categorical data analysis and variance estimation, serving as the basis for tests like goodness-of-fit and independence tests in contingency tables.


18. What is the Chi-square Goodness of Fit Test, and How is it Applied?

The Chi-square test statistic is calculated as:

$$
\chi^2 = \sum_{i=1}^k \frac{(O_i - E_i)^2}{E_i}
$$

Where:  
- $O_i$ = Observed frequency in category $i$  
- $E_i$ = Expected frequency in category $i$ (based on the hypothesized distribution)  
- $k$ = Number of categories

Example Application

Suppose you want to test if a six-sided die is fair:

| Face | Observed Frequency ($O_i$) | Expected Frequency ($E_i$) = Total Rolls / 6 |
|-------|---------------------------|----------------------------------------------|
| 1     | 8                         | 10                                           |
| 2     | 12                        | 10                                           |
| 3     | 9                         | 10                                           |
| 4     | 11                        | 10                                           |
| 5     | 10                        | 10                                           |
| 6     | 10                        | 10                                           |

Calculate:

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} = \frac{(8-10)^2}{10} + \frac{(12-10)^2}{10} + \cdots = 1.2
$$

Compare with critical value for df = 5 at 0.05 significance (11.07). Since 1.2 < 11.07, **do not reject $H_0$** — the die appears fair.


- The Chi-square goodness of fit test evaluates whether observed categorical data fits a specified distribution.
- It is widely used in testing fairness, distribution fitting, and categorical data analysis.


19. What is the F-Distribution, and When Is It Used in Hypothesis Testing?

* It is the distribution of the ratio of two independent chi-square distributed variables divided by their respective degrees of freedom.

- It is **right-skewed** and depends on two sets of degrees of freedom:  
  - Numerator degrees of freedom ($df_1$)  
  - Denominator degrees of freedom ($df_2$)

- The F-distribution is used primarily in tests comparing variances or in analysis of variance (ANOVA).



If $U \sim \chi^2(df_1)$ and $V \sim \chi^2(df_2)$ are independent chi-square variables, then the statistic:

$$
F = \frac{(U / df_1)}{(V / df_2)}
$$

follows an F-distribution with degrees of freedom $df_1$ and $df_2$.

The F-distribution is a fundamental tool in hypothesis testing involving the comparison of variances and means across multiple groups, particularly in ANOVA and regression analysis.



20. What is an ANOVA Test, and What Are Its Assumptions?


**ANOVA (Analysis of Variance)** is a statistical test used to determine whether there are significant differences between the means of three or more independent groups.

- It compares **between-group variance** to **within-group variance**.
- If the between-group variance is significantly larger, we reject the null hypothesis that all group means are equal.

To validly use ANOVA, the following assumptions must be met:

1. **Independence of observations**  
   - The data from different groups should be collected independently.

2. **Normality**  
   - The data in each group should be approximately normally distributed.

3. **Homogeneity of variances (Homoscedasticity)**  
   - The variances in each group should be roughly equal.
   - This can be checked using Levene’s test or Bartlett’s test.

21. What Are the Different Types of ANOVA Tests?


1. One-Way ANOVA

- Use: When comparing means of three or more independent groups based on one independent variable (factor).
- **Example:** Comparing average test scores of students taught using 3 different teaching methods.
- Assumptions: Normality, independence, and equal variances across groups.


2. Two-Way ANOVA (Factorial ANOVA)

- **Use:** When comparing group means based on two independent variables (factors).
- Can detect:
  - Main effects of each factor
  - Interaction effect between the two factors
- Example: Studying how both gender (male/female) and study method (online/in-person) affect test scores.


3. Repeated Measures ANOVA

- **Use:** When the same subjects are measured multiple times under different conditions or over time.
- Controls for individual differences by using within-subject comparisons.
- **Example:** Measuring blood pressure of the same group of patients before, during, and after treatment.

4. Two-Way Repeated Measures ANOVA

- **Use:** When there are two within-subjects factors, or one within-subject and one between-subject factor.
- **Example:** Measuring reaction time of two groups (young vs old) across 3 different lighting conditions.

5. Mixed-Design ANOVA (Split-Plot ANOVA)

- Use: When there is a combination of within-subject and between-subject variables.
- **Example:** One factor (e.g., drug dose) is measured within subjects, and another factor (e.g., gender) is measured between subjects.



22. What is the F-test, and How Does It Relate to Hypothesis Testing?


- It is used to test hypotheses about whether groups differ in terms of **variability** or **overall model fit**.
- The test statistic is the ratio of two variances:
  
  $$
  F = \frac{s_1^2}{s_2^2}
  $$

  where $s_1^2$ and $s_2^2$ are sample variances, with $s_1^2$ usually from the group expected to have a larger variance.



1. **Test for equality of variances**  
   - Compare the variances of two independent samples.
   - Null Hypothesis ($H_0$): $\sigma_1^2 = \sigma_2^2$  
   - Alternative Hypothesis ($H_a$): $\sigma_1^2 \ne \sigma_2^2$

2. **In ANOVA (Analysis of Variance)**  
   - Used to compare **means** across multiple groups by analyzing **variance** between and within groups.
   - Null Hypothesis ($H_0$): All group means are equal.
   - A significant F-statistic indicates **at least one group mean is different**.

3. **In Regression Analysis**  
   - Tests the **overall significance** of the regression model.
   - Null Hypothesis ($H_0$): All regression coefficients = 0 (no predictive power).
   - A significant F-test means the model explains a significant portion of variability in the response variable.



- A **larger F-value** indicates greater disparity between group variances or model fit vs. error.
- Compare the calculated F-statistic to a **critical F-value** (based on degrees of freedom and significance level $\alpha$).
- Alternatively, use the **p-value**:
  - If **p < α**, reject $H_0$
  - If **p ≥ α**, fail to reject $H_0$

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```
import numpy as np
from scipy.stats import norm

# 1. Z-test for comparing a sample mean to a known population mean

# Simulate sample data
sample_mean = 105
population_mean = 100
population_std = 10
sample_size = 30 # Assuming n >= 30 for Z-test

# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
print(f"Calculated Z-statistic: {z_statistic:.4f}")

# Define significance level
alpha = 0.05

# Calculate the critical Z-value for a two-tailed test
critical_z = norm.ppf(1 - alpha/2)
print(f"Critical Z-value (alpha={alpha}): {critical_z:.4f}")

# Calculate the p-value
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
print(f"Calculated P-value: {p_value:.4f}")

# Interpret the results
if abs(z_statistic) > critical_z or p_value < alpha:
    print(f"Interpretation: Reject the null hypothesis (H0). The sample mean ({sample_mean}) is significantly different from the population mean ({population_mean}).")
else:
    print(f"Interpretation: Fail to reject the null hypothesis (H0). The sample mean ({sample_mean}) is not significantly different from the population mean ({population_mean}).")
```

3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```
# 3. Implement a one-sample Z-test using Python
def one_sample_z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05, alternative='two-sided'):
    if sample_size < 30:
        print("Warning: Sample size is less than 30. A T-test might be more appropriate if population standard deviation is unknown.")

    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
    elif alternative == 'less':
        p_value = norm.cdf(z_statistic)
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_statistic)
    else:
        raise ValueError("Invalid 'alternative' argument. Choose from 'two-sided', 'less', or 'greater'.")

    return z_statistic, p_value

# Example usage of the one-sample Z-test function
sample_mean_ex = 110
pop_mean_ex = 100
pop_std_ex = 15
n_ex = 50
alpha_ex = 0.01

z_stat_ex, p_value_ex = one_sample_z_test(sample_mean_ex, pop_mean_ex, pop_std_ex, n_ex, alpha=alpha_ex)

print(f"Sample Mean: {sample_mean_ex}")
print(f"Population Mean (H0): {pop_mean_ex}")
print(f"Population Std: {pop_std_ex}")
print(f"Sample Size: {n_ex}")
print(f"Significance Level (alpha): {alpha_ex}")
print(f"Z-statistic: {z_stat_ex:.4f}")
print(f"P-value: {p_value_ex:.4f}")

if p_value_ex < alpha_ex:
    print("Interpretation: Reject H0.")
else:
    print("Interpretation: Fail to reject H0.")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```
# 4. Perform a two-tailed Z-test and visualize the decision region
import matplotlib.pyplot as plt

def plot_z_test_decision_region(z_statistic, alpha=0.05):
    """
    Visualizes the decision region for a two-tailed Z-test.

    Args:
        z_statistic (float): The calculated Z-statistic.
        alpha (float): The significance level (default is 0.05).
    """
    critical_z = norm.ppf(1 - alpha/2)

    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, y, where=(x > critical_z) | (x < -critical_z), color='red', alpha=0.3, label=f'Rejection Region (α={alpha})')
    plt.axvline(z_statistic, color='blue', linestyle='dashed', linewidth=2, label=f'Z-statistic = {z_statistic:.2f}')
    plt.axvline(critical_z, color='red', linestyle='dotted', linewidth=1.5, label=f'Critical Z = ±{critical_z:.2f}')
    plt.axvline(-critical_z, color='red', linestyle='dotted', linewidth=1.5)

    plt.title('Two-tailed Z-test Decision Region')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
sample_mean_vis = 103
pop_mean_vis = 100
pop_std_vis = 10
n_vis = 40
alpha_vis = 0.05

z_stat_vis = (sample_mean_vis - pop_mean_vis) / (pop_std_vis / np.sqrt(n_vis))
print(f"Calculated Z-statistic: {z_stat_vis:.4f}")

plot_z_test_decision_region(z_stat_vis, alpha_vis)
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors

# This is a conceptual function as true Type II error requires knowing the true effect size.
# We can illustrate the concepts by showing the distributions under H0 and a specific H1.

def visualize_errors(population_mean_h0, population_std, sample_size, alpha, true_mean_h1=None):

    std_error = population_std / np.sqrt(sample_size)

    # Distribution under H0
    x = np.linspace(population_mean_h0 - 4 * std_error, population_mean_h0 + 4 * std_error, 1000)
    y_h0 = norm.pdf(x, population_mean_h0, std_error)

    # Critical values (for a two-tailed test)
    critical_z = norm.ppf(1 - alpha / 2)
    critical_value_upper = population_mean_h0 + critical_z * std_error
    critical_value_lower = population_mean_h0 - critical_z * std_error

    plt.figure(figsize=(12, 7))
    plt.plot(x, y_h0, label='Distribution under H0', color='blue')

    # Type I Error (Alpha)
    plt.fill_between(x, y_h0, where=(x > critical_value_upper) | (x < critical_value_lower), color='red', alpha=0.3, label=f'Type I Error (α={alpha})')
    plt.axvline(critical_value_upper, color='red', linestyle='dashed', linewidth=1.5)
    plt.axvline(critical_value_lower, color='red', linestyle='dashed', linewidth=1.5)
    plt.text(critical_value_upper, plt.gca().get_ylim()[1] * 0.9, f'{critical_value_upper:.2f}', color='red', ha='center')
    plt.text(critical_value_lower, plt.gca().get_ylim()[1] * 0.9, f'{critical_value_lower:.2f}', color='red', ha='center')

    if true_mean_h1 is not None:
        # Distribution under H1
        x_h1 = np.linspace(true_mean_h1 - 4 * std_error, true_mean_h1 + 4 * std_error, 1000)
        y_h1 = norm.pdf(x_h1, true_mean_h1, std_error)
        plt.plot(x_h1, y_h1, label='Distribution under H1', color='green')

        # Type II Error (Beta)
        # The area under H1 between the critical values
        beta_area = norm.cdf(critical_value_upper, true_mean_h1, std_error) - norm.cdf(critical_value_lower, true_mean_h1, std_error)
        plt.fill_between(x_h1, y_h1, where=(x_h1 >= critical_value_lower) & (x_h1 <= critical_value_upper), color='orange', alpha=0.5, label=f'Type II Error (β={beta_area:.4f})')

    plt.title('Type I and Type II Error Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
visualize_errors(population_mean_h0=100, population_std=10, sample_size=30, alpha=0.05)
visualize_errors(population_mean_h0=100, population_std=10, sample_size=30, alpha=0.05, true_mean_h1=105) # Visualize Type II error with a specific H1
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```
from scipy.stats import t

def calculate_confidence_interval(sample_mean, sample_std, sample_size, confidence_level=0.95):

    if sample_size <= 1:
        raise ValueError("Sample size must be greater than 1 to calculate a confidence interval.")

    degrees_of_freedom = sample_size - 1
    alpha_ci = 1 - confidence_level

    # Use t-distribution for confidence interval calculation when population std is unknown
    # and sample size is small. Even with large samples, the t-distribution is generally
    # more robust and approaches the normal distribution as n increases.
    t_critical = t.ppf(1 - alpha_ci / 2, degrees_of_freedom)

    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, margin_of_error

# Example usage
sample_mean_ci = 85
sample_std_ci = 12
sample_size_ci = 20
confidence_level_ci = 0.90

lower_bound_ci, upper_bound_ci, moe_ci = calculate_confidence_interval(sample_mean_ci, sample_std_ci, sample_size_ci, confidence_level_ci)

print(f"Sample Mean: {sample_mean_ci}")
print(f"Sample Std: {sample_std_ci}")
print(f"Sample Size: {sample_size_ci}")
print(f"Confidence Level: {confidence_level_ci}")
print(f"Degrees of Freedom: {sample_size_ci - 1}")
print(f"Margin of Error: {moe_ci:.4f}")
print(f"{int(confidence_level_ci*100)}% Confidence Interval: ({lower_bound_ci:.4f}, {upper_bound_ci:.4f})")

print("\nSignificance of a Confidence Interval:")
print(f"A {int(confidence_level_ci*100)}% confidence interval means that if we were to take many random samples from the same population and calculate a confidence interval for each sample, approximately {int(confidence_level_ci*100)}% of those intervals would contain the true population mean.")
print("It provides a range of plausible values for the population parameter based on our sample data, along with a measure of the uncertainty or precision of our estimate.")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data

```
from scipy.stats import t

def calculate_margin_of_error(sample_std, sample_size, confidence_level=0.95):
    """
    Calculates the margin of error for a sample mean.

    Args:
        sample_std (float): The sample standard deviation.
        sample_size (int): The sample size.
        confidence_level (float): The desired confidence level (default is 0.95).

    Returns:
        float: The margin of error.
    """
    if sample_size <= 1:
        raise ValueError("Sample size must be greater than 1 to calculate margin of error.")

    degrees_of_freedom = sample_size - 1
    alpha = 1 - confidence_level

    # Use t-critical value for margin of error calculation
    t_critical = t.ppf(1 - alpha / 2, degrees_of_freedom)

    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

    return margin_of_error

# Example usage
sample_std_moe = 15
sample_size_moe = 30
confidence_level_moe = 0.99

moe = calculate_margin_of_error(sample_std_moe, sample_size_moe, confidence_level_moe)

print(f"Sample Standard Deviation: {sample_std_moe}")
print(f"Sample Size: {sample_size_moe}")
print(f"Confidence Level: {confidence_level_moe}")
print(f"Calculated Margin of Error: {moe:.4f}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```
# This is a simple conceptual example to illustrate Bayes' Theorem.
# More complex Bayesian inference often involves sampling methods like MCMC.

def bayes_theorem(prior_prob_a, prob_b_given_a, prob_b_given_not_a):
    """
    Applies Bayes' Theorem to calculate the posterior probability.

    Args:
        prior_prob_a (float): Prior probability of hypothesis A, P(A).
        prob_b_given_a (float): Probability of evidence B given hypothesis A is true, P(B|A).
        prob_b_given_not_a (float): Probability of evidence B given hypothesis A is false, P(B|~A).

    Returns:
        float: Posterior probability of hypothesis A given evidence B, P(A|B).
    """
    # Probability of not A
    prior_prob_not_a = 1 - prior_prob_a

    # Probability of B (Marginal Likelihood)
    prob_b = (prob_b_given_a * prior_prob_a) + (prob_b_given_not_a * prior_prob_not_a)

    # Apply Bayes' Theorem
    if prob_b == 0:
        return 0  # Avoid division by zero
    posterior_prob_a_given_b = (prob_b_given_a * prior_prob_a) / prob_b

    return posterior_prob_a_given_b

# Example: Medical Diagnosis
# Hypothesis A: Person has a rare disease (e.g., prevalence = 0.01)
# Evidence B: Person tests positive for the disease

prior_disease = 0.01       # P(Disease)
prob_positive_given_disease = 0.95 # P(Positive | Disease) - Sensitivity of the test
prob_positive_given_no_disease = 0.10 # P(Positive | No Disease) - False positive rate

posterior_disease_given_positive = bayes_theorem(prior_disease, prob_positive_given_disease, prob_positive_given_no_disease)

print(f"Prior Probability of Disease: {prior_disease}")
print(f"Probability of Positive Test given Disease: {prob_positive_given_disease}")
print(f"Probability of Positive Test given No Disease: {prob_positive_given_no_disease}")
print(f"Posterior Probability of Disease given Positive Test: {posterior_disease_given_positive:.4f}")

print("\nExplanation of Bayesian Inference Process:")
print("1. Start with a Prior Probability (P(A)): This is your initial belief about the likelihood of the hypothesis before observing any new evidence.")
print("2. Gather Evidence (B): Obtain new data or observations.")
print("3. Calculate the Likelihood (P(B|A)): Determine the probability of observing the evidence, assuming your hypothesis is true.")
print("4. Calculate the Marginal Likelihood (P(B)): Determine the total probability of observing the evidence, considering all possible hypotheses.")
print("5. Apply Bayes' Theorem to Calculate the Posterior Probability (P(A|B)): Update your belief about the hypothesis based on the evidence.")
print("   P(A|B) = [P(B|A) * P(A)] / P(B)")
print("This process allows us to update our beliefs in a rational way as we accumulate more evidence.")
```

```
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate data for a two-way ANOVA
# Factors: Factor A (e.g., Treatment: Drug A, Drug B, Placebo), Factor B (e.g., Gender: Male, Female)
# Dependent variable: Response (e.g., Blood Pressure Reduction)

np.random.seed(42)

# Generate data for each combination of factors
n_per_group = 20
data_a_male = np.random.normal(10, 3, n_per_group)
data_a_female = np.random.normal(12, 3, n_per_group)
data_b_male = np.random.normal(15, 3, n_per_group)
data_b_female = np.random.normal(14, 3, n_per_group)
data_placebo_male = np.random.normal(5, 3, n_per_group)
data_placebo_female = np.random.normal(6, 3, n_per_group)

# Create a Pandas DataFrame
data_2way = pd.DataFrame({
    'Response': np.concatenate([data_a_male, data_a_female, data_b_male, data_b_female, data_placebo_male, data_placebo_female]),
    'Treatment': ['Drug A'] * n_per_group * 2 + ['Drug B'] * n_per_group * 2 + ['Placebo'] * n_per_group * 2,
    'Gender': ['Male'] * n_per_group + ['Female'] * n_per_group + ['Male'] * n_per_group + ['Female'] * n_per_group + ['Male'] * n_per_group + ['Female'] * n_per_group
})

# Perform two-way ANOVA using statsmodels
# The formula specifies the model: Response is predicted by Treatment, Gender, and their interaction (Treatment * Gender)
model = ols('Response ~ C(Treatment) + C(Gender) + C(Treatment):C(Gender)', data=data_2way).fit()
anova_table = sm.stats.anova_lm(model, typ=2) # typ=2 for Type II ANOVA

print("Two-Way ANOVA Results:")
print(anova_table)

# Interpret the results from the ANOVA table:
# Look at the 'PR(>F)' column (p-value) for each row:
# - C(Treatment): Significance of the main effect of Treatment.
# - C(Gender): Significance of the main effect of Gender.
# - C(Treatment):C(Gender): Significance of the interaction effect between Treatment and Gender.

alpha_2way_anova = 0.05

print("\nInterpretation:")
if anova_table['PR(>F)']['C(Treatment):C(Gender)'] < alpha_2way_anova:
    print("  - Interaction effect (Treatment * Gender) is statistically significant.")
    print("    The effect of one factor depends on the level of the other factor.")
elif anova_table['PR(>F)']['C(Treatment)'] < alpha_2way_anova or anova_table['PR(>F)']['C(Gender)'] < alpha_2way_anova:
     print("  - Interaction effect is NOT significant, but at least one main effect is significant.")
     if anova_table['PR(>F)']['C(Treatment)'] < alpha_2way_anova:
         print("    - Main effect of Treatment is significant.")
     if anova_table['PR(>F)']['C(Gender)'] < alpha_2way_anova:
         print("    - Main effect of Gender is significant.")
else:
    print("  - Neither interaction nor main effects are statistically significant.")


# Visualize the interaction effect (often with an interaction plot)
plt.figure(figsize=(10, 6))
sns.pointplot(x='Treatment', y='Response', hue='Gender', data=data_2way, capsize=.1)
plt.title('Interaction Plot: Effect of Treatment and Gender on Response')
plt.xlabel('Treatment')
plt.ylabel('Response')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Visualize the main effects (e.g., boxplots or bar plots for each factor separately)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='Treatment', y='Response', data=data_2way)
plt.title('Main Effect of Treatment')
plt.xlabel('Treatment')
plt.ylabel('Response')
plt.grid(True, linestyle='--', alpha=0.6)

plt.subplot(1, 2, 2)
sns.boxplot(x='Gender', y='Response', data=data_2way)
plt.title('Main Effect of Gender')
plt.xlabel('Gender')
plt.ylabel('Response')
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

```
from scipy.stats import ttest_1samp, norm
import numpy as np

# Simulate data from a normal distribution
np.random.seed(42)
sample_data_hypothesis = np.random.normal(loc=55, scale=10, size=40) # Sample mean ~55, Std Dev 10, size 40

# Hypothesize a population mean
hypothesized_population_mean = 50

print(f"Simulated Sample Mean: {np.mean(sample_data_hypothesis):.4f}")
print(f"Hypothesized Population Mean (H0): {hypothesized_population_mean}")

# Perform a one-sample T-test (appropriate when population standard deviation is unknown, which is common with simulated data unless explicitly set)
# Null Hypothesis (H0): The true population mean is equal to the hypothesized mean (50).
# Alternative Hypothesis (H1): The true population mean is not equal to the hypothesized mean (50).
t_statistic_hypothesis, p_value_hypothesis = ttest_1samp(sample_data_hypothesis, hypothesized_population_mean)

print(f"\nCalculated T-statistic: {t_statistic_hypothesis:.4f}")
print(f"Calculated P-value: {p_value_hypothesis:.4f}")

alpha_hypothesis = 0.05

# Interpret the results
if p_value_hypothesis < alpha_hypothesis:
    print(f"Interpretation: Reject the null hypothesis (H0). The sample mean is statistically significantly different from the hypothesized population mean of {hypothesized_population_mean}.")
else:
    print(f"Interpretation: Fail to reject the null hypothesis (H0). The sample mean is not statistically significantly different from the hypothesized population mean of {hypothesized_population_mean}.")
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```
from scipy.stats import chi2
import numpy as np

# Simulate sample data
np.random.seed(42)
sample_data_variance = np.random.normal(50, 8, 25) # Sample from a population with variance 8^2 = 64
sample_variance = np.var(sample_data_variance, ddof=1) # Calculate sample variance (unbiased)

# Hypothesize a population variance
hypothesized_population_variance = 60

sample_size_variance = len(sample_data_variance)
degrees_of_freedom_variance = sample_size_variance - 1

# Calculate the Chi-square test statistic for variance
# Formula: chi2_stat = (n - 1) * s^2 / sigma^2
# where n is sample size, s^2 is sample variance, sigma^2 is hypothesized population variance
chi2_statistic_variance = (degrees_of_freedom_variance * sample_variance) / hypothesized_population_variance

print(f"Sample Variance: {sample_variance:.4f}")
print(f"Hypothesized Population Variance (H0): {hypothesized_population_variance}")
print(f"Sample Size: {sample_size_variance}")
print(f"Degrees of Freedom: {degrees_of_freedom_variance}")
print(f"Calculated Chi-square statistic (Variance Test): {chi2_statistic_variance:.4f}")

# Calculate the p-value for a two-tailed test
# We need the probability in both tails of the Chi-square distribution.
# For a two-tailed test, the p-value is 2 * min(P(Chi2 <= observed_chi2), P(Chi2 >= observed_chi2)).
p_value_variance = 2 * min(chi2.cdf(chi2_statistic_variance, degrees_of_freedom_variance),
                           1 - chi2.cdf(chi2_statistic_variance, degrees_of_freedom_variance))

print(f"Calculated P-value (Variance Test - Two-tailed): {p_value_variance:.4f}")

alpha_variance = 0.05

# Interpret the results
if p_value_variance < alpha_variance:
    print(f"Interpretation: Reject the null hypothesis (H0). The sample variance ({sample_variance:.4f}) is statistically significantly different from the hypothesized population variance of {hypothesized_population_variance}.")
else:
    print(f"Interpretation: Fail to reject the null hypothesis (H0). The sample variance ({sample_variance:.4f}) is not statistically significantly different from the hypothesized population variance of {hypothesized_population_variance}.")
```

```
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

# Simulate data for two groups (number of successes and sample size)
# Example: Comparing success rates in two different marketing campaigns
successes1 = 50  # Number of clicks in Campaign 1
nobs1 = 100    # Total emails sent in Campaign 1

successes2 = 65  # Number of clicks in Campaign 2
nobs2 = 120    # Total emails sent in Campaign 2

# Perform the Z-test for two proportions
# Null Hypothesis (H0): The proportions are equal (p1 = p2).
# Alternative Hypothesis (H1): The proportions are not equal (p1 != p2) (two-sided test).
z_statistic_prop, p_value_prop = proportions_ztest([successes1, successes2], [nobs1, nobs2])

proportion1 = successes1 / nobs1
proportion2 = successes2 / nobs2

print(f"Proportion in Group 1: {proportion1:.4f} ({successes1}/{nobs1})")
print(f"Proportion in Group 2: {proportion2:.4f} ({successes2}/{nobs2})")
print(f"\nCalculated Z-statistic (Proportions Test): {z_statistic_prop:.4f}")
print(f"Calculated P-value (Proportions Test): {p_value_prop:.4f}")

alpha_prop = 0.05

# Interpret the results
if p_value_prop < alpha_prop:
    print(f"Interpretation: Reject the null hypothesis (H0). There is a statistically significant difference between the proportions of the two groups.")
else:
    print(f"Interpretation: Fail to reject the null hypothesis (H0). There is no statistically significant difference between the proportions of the two groups.")
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

```
from scipy.stats import f
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate data for two samples (same as Question 16)
np.random.seed(42)
sample1_fvis = np.random.normal(50, 10, 20) # Mean 50, Std Dev 10, size 20
sample2_fvis = np.random.normal(55, 15, 25) # Mean 55, Std Dev 15, size 25

# Calculate sample variances
variance1_fvis = np.var(sample1_fvis, ddof=1)
variance2_fvis = np.var(sample2_fvis, ddof=1)

print(f"Variance of Sample 1: {variance1_fvis:.4f}")
print(f"Variance of Sample 2: {variance2_fvis:.4f}")

# Calculate the F-statistic
if variance1_fvis >= variance2_fvis:
    f_statistic_fvis = variance1_fvis / variance2_fvis
    df1_fvis = len(sample1_fvis) - 1
    df2_fvis = len(sample2_fvis) - 1
else:
    f_statistic_fvis = variance2_fvis / variance1_fvis
    df1_fvis = len(sample2_fvis) - 1
    df2_fvis = len(sample1_fvis) - 1

# Calculate the p-value (two-tailed)
p_value_fvis = 2 * (1 - f.cdf(f_statistic_fvis, df1_fvis, df2_fvis))

print(f"\nCalculated F-statistic (Variance Test): {f_statistic_fvis:.4f}")
print(f"Degrees of Freedom (Numerator, Denominator): ({df1_fvis}, {df2_fvis})")
print(f"Calculated P-value (Variance Test - Two-tailed): {p_value_fvis:.4f}")

alpha_fvis = 0.05

# Interpret the results
if p_value_fvis < alpha_fvis:
    print(f"Interpretation: Reject the null hypothesis (H0). There is a statistically significant difference between the variances of the two samples.")
else:
    print(f"Interpretation: Fail to reject the null hypothesis (H0). There is no statistically significant difference between the variances of the two samples.")

# Visualize the F-distribution and the calculated F-statistic
# Create an F-distribution based on the degrees of freedom
x_f_vis = np.linspace(0, f.ppf(0.999, df1_fvis, df2_fvis), 1000) # Generate points up to a high percentile
y_f_vis = f.pdf(x_f_vis, df1_fvis, df2_fvis)

plt.figure(figsize=(10, 6))
plt.plot(x_f_vis, y_f_vis, label=f'F-distribution (df1={df1_fvis}, df2={df2_fvis})')

# Calculate critical F-value for a two-tailed test (alpha/2 in each tail)
# Since our calculated F-statistic is always >= 1, we only need the upper critical value for the visualization
critical_f_upper = f.ppf(1 - alpha_fvis/2, df1_fvis, df2_fvis)
# The lower critical value would be f.ppf(alpha_fvis/2, df1_fvis, df2_fvis)

# Shade the rejection region(s)
# Since we put the larger variance in the numerator, we are primarily interested in the upper tail for rejection
plt.fill_between(x_f_vis, y_f_vis, where=(x_f_vis > critical_f_upper), color='red', alpha=0.3, label=f'Rejection Region (α/2={alpha_fvis/2})')
# For a full two-tailed test, you'd also shade the lower tail, but with the F-ratio >= 1 logic, the lower tail is less relevant for the calculated statistic.

plt.axvline(f_statistic_fvis, color='blue', linestyle='dashed', linewidth=2, label=f'Calculated F-statistic = {f_statistic_fvis:.2f}')
plt.axvline(critical_f_upper, color='red', linestyle='dotted', linewidth=1.5, label=f'Critical F (upper) = {critical_f_upper:.2f}')

plt.title('F-test for Comparing Variances: F-distribution and Decision Region')
plt.xlabel('F-value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(0) # F-values are non-negative
plt.ylim(0)
plt.show()

# Also visualize the distribution of the two samples
plt.figure(figsize=(8, 6))
sns.histplot(sample1_fvis, kde=True, color='skyblue', label='Sample 1', stat='density')
sns.histplot(sample2_fvis, kde=True, color='lightcoral', label='Sample 2', stat='density')
plt.title('Distribution of the Two Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
